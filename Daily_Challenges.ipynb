## Challenge 1: Avalanche Risk Analysis
**Goal:** Given snow depth and slope, determine if there is an avalanche risk.

he snow depth values are "Shallow", "Moderate", or "Deep".
Slope values are "Gentle", "Steep", or "Very Steep".
Return "Safe" or "Risky" based on this table:

| Slope | Shallow | Moderate | Deep |
|---|---|---|---|
| Gentle | Safe | Safe | Safe |
| Steep | Safe | Risky | Risky |
| Very Steep | Safe | Risky | Risky |

```python
# Challenge 1: Avalanche Risk

def avalanche_risk(snow_depth, slope):
    # Logic: It is "Safe" only if the snow is Shallow OR the slope is Gentle.
    # Otherwise, it's "Risky".
    if snow_depth == "Shallow" or slope == "Gentle":
        return "Safe"
    else:
        return "Risky"

# --- TEST AREA ---
# This part runs the code so you can see the result below the cell
print("Testing Shallow/Steep:", avalanche_risk("Shallow", "Steep"))     # Expected: Safe
print("Testing Deep/Gentle:", avalanche_risk("Deep", "Gentle"))        # Expected: Safe
print("Testing Moderate/Steep:", avalanche_risk("Moderate", "Steep"))  # Expected: Risky
```

# üèÜ 2026 Winter Games Day 2: Digital Detox
## Challenge 2: Device Log Verification
> **Goal:** Analyze login logs to determine if a digital detox goal was achieved based on time-based constraints.

### üìã The Rules
To meet the goal, **both** of the following must be true:
1.  **4-Hour Rule:** You logged in no more than once within any four-hour period.
2.  **Daily Limit:** You logged in no more than 2 times on any single day.

### üì• Input Format
* An array (list) of strings.
* Format: `"YYYY-MM-DD HH:mm:ss"` (e.g., `"2026-02-19 14:30:00"`)

### üì§ Expected Output
* `True` if both goals are met.
* `False` if either rule is broken.

---

```python
from datetime import datetime, timedelta

def digital_detox(logs):
    # 1. Convert strings to datetime objects and sort them
    times = sorted([datetime.strptime(log, "%Y-%m-%d %H:%M:%S") for log in logs])

    # 2. Check the 4-hour rule
    for i in range(1, len(times)):
        if times[i] - times[i-1] < timedelta(hours=4):
            return False # Too close together!

    # 3. Check the "Max 2 per day" rule
    dates = [t.date() for t in times]
    for d in dates:
        if dates.count(d) > 2:
            return False # Too many logins in one day!

    return True # If it passed both checks!
```

```python
# --- TEST SUITE ---

# Case 1: PASS (2 logins, 6 hours apart)
logs1 = ["2026-02-19 08:00:00", "2026-02-19 14:00:00"]
print("Test 1 (Should be True):", digital_detox(logs1))

# Case 2: FAIL (2 logins, but only 2 hours apart)
logs2 = ["2026-02-19 08:00:00", "2026-02-19 10:00:00"]
print("Test 2 (Should be False - Too close):", digital_detox(logs2))

# Case 3: FAIL (3 logins in one day)
logs3 = ["2026-02-20 08:00:00", "2026-02-20 13:00:00", "2026-02-20 18:00:00"]
print("Test 3 (Should be False - Too many in a day):", digital_detox(logs3))
```

# üèÜ 2026 Winter Games Day 3: Biathlon
## Challenge 3: Penalty Distance Calculation
**Goal:** Calculate the total penalty distance in meters for all rounds.
* **Input:** An array (list) of integers representing hits per round.
* **Rules:** * 5 targets exist per round.
    * Each miss = 150 meter penalty loop.
* **Logic:** `Total Distance = (Total Misses) * 150`

---

```python
def calculate_penalty_distance(rounds):
    # This variable keeps track of the total distance
    total_penalty_meters = 0

    for hits in rounds:
        # Step 1: Find how many targets were missed (5 is the total)
        misses = 5 - hits

        # Step 2: Multiply misses by 150 to get meters for this round
        round_distance = misses * 150

        # Step 3: Add this round to our total tally
        total_penalty_meters += round_distance

    return total_penalty_meters
```

```python
# --- TEST CASES ---
# Test 1: Round 1 (4 hits = 1 miss), Round 2 (3 hits = 2 misses). Total 3 misses.
print("Test 1 Result:", calculate_penalty_distance([4, 3]), "meters")

# Test 2: Perfect shooting (5 hits each). Should be 0.
print("Test 2 Result:", calculate_penalty_distance([5, 5]), "meters")
```
