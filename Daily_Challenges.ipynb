# 🏆 2026 Winter Games Day 1: Ski Mountaineering
## Challenge 1: Avalanche Risk Analysis
**Goal:** Given snow depth and slope, determine if there is an avalanche risk.

he snow depth values are "Shallow", "Moderate", or "Deep".
Slope values are "Gentle", "Steep", or "Very Steep".
Return "Safe" or "Risky" based on this table:

| Slope | Shallow | Moderate | Deep |
|---|---|---|---|
| Gentle | Safe | Safe | Safe |
| Steep | Safe | Risky | Risky |
| Very Steep | Safe | Risky | Risky |

```python
# Challenge 1: Avalanche Risk

def avalanche_risk(snow_depth, slope):
    # Logic: It is "Safe" only if the snow is Shallow OR the slope is Gentle.
    # Otherwise, it's "Risky".
    if snow_depth == "Shallow" or slope == "Gentle":
        return "Safe"
    else:
        return "Risky"

# --- TEST AREA ---
# This part runs the code so you can see the result below the cell
print("Testing Shallow/Steep:", avalanche_risk("Shallow", "Steep"))     # Expected: Safe
print("Testing Deep/Gentle:", avalanche_risk("Deep", "Gentle"))        # Expected: Safe
print("Testing Moderate/Steep:", avalanche_risk("Moderate", "Steep"))  # Expected: Risky
```

# 🏆 2026 Winter Games Day 2: Digital Detox
## Challenge 2: Device Log Verification
> **Goal:** Analyze login logs to determine if a digital detox goal was achieved based on time-based constraints.

### 📋 The Rules
To meet the goal, **both** of the following must be true:
1.  **4-Hour Rule:** You logged in no more than once within any four-hour period.
2.  **Daily Limit:** You logged in no more than 2 times on any single day.

### 📥 Input Format
* An array (list) of strings.
* Format: `"YYYY-MM-DD HH:mm:ss"` (e.g., `"2026-02-19 14:30:00"`)

### 📤 Expected Output
* `True` if both goals are met.
* `False` if either rule is broken.

---

```python
from datetime import datetime, timedelta

def digital_detox(logs):
    # 1. Convert strings to datetime objects and sort them
    times = sorted([datetime.strptime(log, "%Y-%m-%d %H:%M:%S") for log in logs])

    # 2. Check the 4-hour rule
    for i in range(1, len(times)):
        if times[i] - times[i-1] < timedelta(hours=4):
            return False # Too close together!

    # 3. Check the "Max 2 per day" rule
    dates = [t.date() for t in times]
    for d in dates:
        if dates.count(d) > 2:
            return False # Too many logins in one day!

    return True # If it passed both checks!
```

```python
# --- TEST SUITE ---

# Case 1: PASS (2 logins, 6 hours apart)
logs1 = ["2026-02-19 08:00:00", "2026-02-19 14:00:00"]
print("Test 1 (Should be True):", digital_detox(logs1))

# Case 2: FAIL (2 logins, but only 2 hours apart)
logs2 = ["2026-02-19 08:00:00", "2026-02-19 10:00:00"]
print("Test 2 (Should be False - Too close):", digital_detox(logs2))

# Case 3: FAIL (3 logins in one day)
logs3 = ["2026-02-20 08:00:00", "2026-02-20 13:00:00", "2026-02-20 18:00:00"]
print("Test 3 (Should be False - Too many in a day):", digital_detox(logs3))
```

# 🏆 2026 Winter Games Day 3: Biathlon
## Challenge 3: Penalty Distance Calculation
**Goal:** Calculate the total penalty distance in meters for all rounds.
* **Input:** An array (list) of integers representing hits per round.
* **Rules:** * 5 targets exist per round.
    * Each miss = 150 meter penalty loop.
* **Logic:** `Total Distance = (Total Misses) * 150`

---

```python
def calculate_penalty_distance(rounds):
    # This variable keeps track of the total distance
    total_penalty_meters = 0

    for hits in rounds:
        # Step 1: Find how many targets were missed (5 is the total)
        misses = 5 - hits

        # Step 2: Multiply misses by 150 to get meters for this round
        round_distance = misses * 150

        # Step 3: Add this round to our total tally
        total_penalty_meters += round_distance

    return total_penalty_meters
```

```python
# --- TEST CASES ---
# Test 1: Round 1 (4 hits = 1 miss), Round 2 (3 hits = 2 misses). Total 3 misses.
print("Test 1 Result:", calculate_penalty_distance([4, 3]), "meters")

# Test 2: Perfect shooting (5 hits each). Should be 0.
print("Test 2 Result:", calculate_penalty_distance([5, 5]), "meters")
```

# 🏆 2026 Winter Games Day 4: Ski Jumping
## Challenge 4: Medal Analysis

**Goal:** Determine which medal a jumper receives based on their score relative to others.
* **Input:** `distance_points`, `style_points`, `wind_comp`, `k_point_bonus`
* **Rules:**
    * Combine points into a `total` score.
    * Compare against other competitors' scores: `[165.5, 172.0, 158.0, 180.0, 169.5, 175.0, 162.0, 170.0]`
    * Rank 1 (Index 0) = Gold, Rank 2 (Index 1) = Silver, Rank 3 (Index 2) = Bronze.
* **Logic & Skills:** * **List Concatenation:** Combining your score with the group.
    * **Descending Sort:** Using `.sort(reverse=True)` to put high scores first.
    * **Indexing:** Using `.index()` to find the athlete's rank.

---

```python
def ski_jump_medal(distance_points, style_points, wind_comp, k_point_bonus):
    # Calculate total score
    total = distance_points + style_points + wind_comp + k_point_bonus

    # List of existing competitor scores
    scores = [165.5, 172.0, 158.0, 180.0, 169.5, 175.0, 162.0, 170.0]

    # Add your score to the list
    all_scores = scores + [total]

    # Sort High-to-Low so index 0 is the winner
    all_scores.sort(reverse=True)

    # Find your position (rank) in the list
    rank = all_scores.index(total)

    # Assign medals based on rank
    if rank == 0:
        return "Gold"
    elif rank == 1:
        return "Silver"
    elif rank == 2:
        return "Bronze"
    else:
        return "No Medal"

# --- TEST CASE ---
print("Result:", ski_jump_medal(125.0, 58.0, 0.0, 6.0)) # Expected: Gold
```

🏅 Day 5: Cross-Country Skiing
Topic: String Parsing & Time Delta Logic

📖 The Mission
In long-distance races like Cross-Country Skiing, the official results often show the winner's total time, but for everyone else, they show the time behind the winner (the "delta").

Your Task:

Input: A list of times in "H:MM:SS" format (fastest to slowest).

Output:

The winner’s time becomes "0".

Other times become "+M:SS" (minutes and seconds behind the winner).

💻 The Logic
Your implementation uses the "Normalize to Seconds" strategy, which is the gold standard for time math in Python.

```python
  def get_relative_results(results):
    def to_total_seconds(time_str):
        # Step 1: Split "H:MM:SS" and convert to total seconds for easy math
        h, m, s = map(int, time_str.split(':'))
        return h * 3600 + m * 60 + s

    if not results:
        return []

    # Step 2: Set the baseline (the winner's time)
    winner_seconds = to_total_seconds(results[0])
    output = ["0"]

    # Step 3: Calculate the 'Delta' for the rest
    for time_str in results[1:]:
        current_seconds = to_total_seconds(time_str)
        delta_seconds = current_seconds - winner_seconds

        # Convert delta back to M:SS format
        m = delta_seconds // 60
        s = delta_seconds % 60

        # f-string :02d ensures 5 seconds becomes "05"
        output.append(f"+{m}:{s:02d}")

    return output

# --- TEST CASE ---
race_times = ["1:25:32", "1:26:10", "1:27:05"]
print(f"Original Times: {race_times}")
print(f"Relative Times: {get_relative_results(race_times)}")
```

❄️ 2026 Winter Games Python Challenge: Figure Skater's Final Score

Challenge: Compute the Skater's Final Score

Goal: Calculate a figure skater’s final competition score by removing biased judge scores and applying penalties.

Input:

* A list `scores` containing **exactly 10 judge scores** (each between **0.0 and 10.0**).
* Optional penalty values representing technical violations (falls, overtime, etc.).

Rules:

* Remove the **highest** and **lowest** score to eliminate possible bias.
* Sum the remaining **8 scores** to form the base score.
* Subtract **any penalties** provided from the base score.
* Return the **final score**.

Example:
Input:
`scores = [9.0, 8.5, 9.2, 8.8, 9.5, 9.1, 8.7, 9.3, 8.9, 9.4]`
Penalties: `0.5`

Process:

* Highest score → `9.5` (removed)
* Lowest score → `8.5` (removed)
* Remaining scores summed → base score
* Subtract penalty → final score

Logic & Skills:

* **Sorting Lists:** Use `sorted()` or `.sort()` to identify highest and lowest values.
* **List Slicing:** Extract the middle 8 scores after sorting.
* **Variable Arguments:** Use `*penalties` to accept multiple optional penalty values.
* **Aggregation Pattern:** Use `sum()` to compute totals.



```python
def compute_score(judge_scores, *penalties):
    """
    Calculates final figure skating score by removing outliers
    and applying optional technical penalties.
    """
    # Remove max and min to get the middle 8 scores
    base_score = sum(judge_scores) - max(judge_scores) - min(judge_scores)

    # Subtract the sum of all penalty arguments
    final_score = base_score - sum(penalties)

    # Return 0 if penalties exceed the score
    return max(0, final_score)

# Example Test Case
print(compute_score([8, 9, 8.5, 7, 10, 6, 8, 9, 8.5, 9], 1.0, 0.5))
```

🏆 Python Practice Challenge: Sum the Letters

Challenge: Letter Value Sum

Goal: Calculate the sum of letter values in a string.

Input:
A string `s` containing letters, numbers, spaces, or symbols.

Rules:

* Letters are **A–Z or a–z**.
* Letter values follow:

  * **A = 1, B = 2, C = 3 ... Z = 26**
* Uppercase and lowercase letters have the **same value**.
* Ignore **all non-letter characters**.

Example:
Input: `"aB!"`

Process:

* `a → 1`
* `B → 2`
* `! → ignored`

Output:
`3`

Logic & Skills:

* **String Iteration:** Loop through each character using `for char in s`.
* **Character Checking:** Use `.isalpha()` to detect letters.
* **ASCII Conversion:** Use `ord(char.upper()) - 64` to convert letters to values.
* **Accumulation Pattern:** Use a `total` variable to keep adding values.



```python
def sum_letters(s):
    total = 0

    for char in s:
        if char.isalpha():
            value = ord(char.upper()) - 64
            total += value

    return total

# Example Test:
print(sum_letters("aB!"))   # Output: 3
print(sum_letters("Hello")) # Output: 52
```
